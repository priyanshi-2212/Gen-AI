Loops control statement
break statement--this statement break the loop
continue statement--skip the currrent iteration
pass statement--only passed the code

```python
i = 0
while i < 5:
  i += 1
  if i == 3:
    break
  else:  # else should be aligned with if
    print(i)
```

```python
i = 0
while i<5:
  i+=1
  if i==3:
    continue
    print(i)
  else:
    print(i)
```

```python
i = 0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
    print(i)
```

variable length arguement--we give any number of arguement

1. Non keyword variable length arguement--it is behave like tuple, represent by star(*)

2. Keyword variable length arguement--it is behave like dictionary, represent by double star(**)

```python
# Non keyword variable length arguement
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
 test(1, 2, 3, 4, 5, 6)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(1, 2, 3, 4, 5, 6, 7, 8, 9)
```

```python
# keyword variable length arguement
def test(**kwargs):
    print(kwargs)
    print(len(kwargs))
    print(type(kwargs))
```

```python
test(a = 10, b = 20, c = 30)
```

```python
import numpy as np
import matplotlib.pyplot as plt
cgpa = np.array([6, 7, 8, 9,10])
package = np.array([4, 5, 6, 7, 8])
plt.plot(cgpa, package, "*:g", ms= 16, mec="pink" , mfc = "blue", linewidth=8)
plt.xlabel("CGPA")
plt.ylabel("Package in LPA")
plt.title("Shoolini University Students data")
plt.grid()
plt.show()
```

```python
#bar plot--it work on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "banana", "Orange", "guava"])
quantity = np.array([300, 400,200,50])
color = np.array(["Red", "Yellow", "Orange", "Green"])
plt.bar(fruit, quantity,color = color)
plt.xlabel("fruit")
plt.ylabel("quantity")
plt.title("fruit vs fruit quantity")
plt.show()

```

```python
#pie plot

percentage_of_fruit= np.array([20, 25, 30, 25])
fruit = np.array(["Apple", "banana", "Orange", "guava"])
myexplode= np.array([0.2,0, 0,0])
plt.pie(percentage_of_fruit, labels = fruit, explode = myexplode)
plt.legend(title = "fruit", loc = "best")
plt.show()
```
